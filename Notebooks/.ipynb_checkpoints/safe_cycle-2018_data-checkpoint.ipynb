```python
crash_df = pd.read_csv('../data/NYPD_Motor_Vehicle_Collisions-June2019.csv')
```

```python
def remove_col_spaces(df):
    df.columns = df.columns.map(lambda x: x.lower().replace(' ','_'))
    return None

def rename_cols(df):
    '''Renames columns'''
    new_col_dict = {'number_of_persons_injured': 'num_pers_i','number_of_persons_killed': 'num_pers_k', 
                    'number_of_pedestrians_injured': 'num_ped_i', 'number_of_pedestrians_killed': 'num_ped_k',
                    'number_of_cyclist_injured': 'num_c_i', 'number_of_cyclist_killed': 'num_c_k',
                    'number_of_motorist_injured': 'num_m_i', 'number_of_motorist_killed': 'num_m_k',
                    'contributing_factor_vehicle_1': 'cf_1', 'contributing_factor_vehicle_2': 'cf_2',
                    'contributing_factor_vehicle_3': 'cf_3', 'contributing_factor_vehicle_4': 'cf_4',
                    'contributing_factor_vehicle_5': 'cf_5',
                    'vehicle_type_code_1': 'veh_type_1', 'vehicle_type_code_2': 'veh_type_2', 
                    'vehicle_type_code_3': 'veh_type_3', 'vehicle_type_code_4': 'veh_type_4', 
                    'vehicle_type_code_5': 'veh_type_5',
                   }
    df.rename(columns = new_col_dict,inplace=True)
    return None

def fix_date_time(df):
    '''Combines date and time columns into a single datetime column'''
    try: isinstance(crash_df['date_time'],pd.Series)
    except KeyError:   
        locs = crash_df[crash_df['time'].map(lambda x: x.strip()[1] == ':')].index
        crash_df.at[locs,'time'] = '0' + crash_df.loc[locs]['time']
        test_ser = crash_df['date'].astype(str) + ' ' + crash_df['time'] + ':00'
        crash_df['date_time'] = pd.to_datetime(test_ser,format='%m/%d/%Y %X')
        crash_df.drop(['date','time'],axis=1,inplace=True)
    return None

def fix_location(df):
    '''Changes type of location tuple entries from string to float'''
    df['location'] = list(zip(df['latitude'],df['longitude']))
    return None
```

```python
remove_col_spaces(crash_df)
rename_cols(crash_df)
fix_date_time(crash_df)
fix_location(crash_df)
```

```python
def dist_sq(a, b):
    return (a[0] - b[0])**2 + .41*(a[1] - b[1])**2 #.41 ~ sin(Pi/4))**2

def get_closest_int(b):
    return np.argmin([dist_sq(a[2:],b) for a in int_list])

def latlon_dist_sq(df, latlon_tup):
    return (df['latitude'] - latlon_tup[0])**2 + .425*(df['longitude'] - latlon_tup[1])**2 #.41 ~ sin(Pi/4))**2
```

```python
def get_speed_limits(limits_df, int_df):
    
    int_df['sum_speed_limit'] = np.zeros(int_df['coord'].shape)
    
    int_df['num_rds'] = np.zeros(int_df['coord'].shape)
    
    limits_df['lines'] = limits_df['the_geom'].map(lambda x: loads(x))
    
    for ix in limits_df.index:
        multi = limits_df.loc[ix]['lines']
        speed = limits_df.loc[ix]['postvz_sl']
        
        locs = int_df[int_df['pt'].map(lambda x: multi.distance(x) < .0001)].index
        int_df.loc[locs,'sum_speed_limit'] += speed #int_df.loc[locs]['bike_lanes'].map(lambda x: x + num_lanes)
        int_df.loc[locs,'num_rds'] += 1#int_df.loc[locs]['bike_lane_type'].map(lambda x: x + ',' + lane_types)
    return int_df
```

```python
def get_bumps(hump_df, int_df):
    
    int_df['speed_bumps'] = list(zip(np.zeros(int_df['coord'].shape),np.zeros(int_df['coord'].shape))) #need this to be an array of tuples if want to add in date
    
    for ix in hump_df.index:
        multi = hump_df.loc[ix]['lines']
        num_bumps = hump_df.loc[ix]['new_humps']
        date = hump_df.loc[ix]['date']
        locs = int_df[int_df['pt'].map(lambda x: multi.distance(x) < .0001)].index
        int_df.at[locs,'speed_bumps'] = int_df.loc[locs]['speed_bumps'].map(lambda x: (x[0] + num_bumps, date) )
        
    return int_df
```

```python
def get_line_feature(feature_df, int_df, feat_type):
    
    int_df[feat_type] = list(zip(np.zeros(int_df['coord'].shape),np.zeros(int_df['coord'].shape)))
    
    feature_df['lines'] = feature_df['the_geom'].map(lambda x: loads(x))
    
    for ix in feature_df.index:
        multi = feature_df.loc[ix]['lines']
        if feat_type in []:
            num_feat = hump_df.loc[ix]['']
        else:
            num_feat = 1
        date = feature_df.loc[ix]['ENFORCEMEN']
        locs = int_df[int_df['pt'].map(lambda x: multi.distance(x) < .0001)].index
        int_df.at[locs,feat_type] = int_df.loc[locs][feat_type].map(lambda x: (x[0] + num_feat, date) )
        
    return int_df
```
